```r
#Let's create a function for evaluating results 
#The function takes true and predicted as 

evaluate_results <- function(true, predicted, df) {
    SSE <- ((true-predicted)^2)
    SST <- ((true-mean(true))^2)
    R2 <- 1 - SSE/SST
    RMSE <- sqrt(SSE/nrow(df))
            
    data.frame(
        RMSE = RMSE, 
        R_squared = R2)
    
}
```
